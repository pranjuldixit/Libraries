```python
import pandas as pd
```

```python
import numpy as np
```

```python

# Load dataset
df = pd.read_csv("G:\\Pw skills\\Assignment 37(Python)\\retail_dataset .csv")
```

```python
print("===== Q1: Dataset Preview & Info =====")
print(df.head())
print(df.info())
```

```python
print("\n===== Q2: Missing Values =====")
missing = df.isnull().sum()
print(missing)
print("Total Missing Values:", missing.sum())
```

```python
print("\n===== Q3: Fill Missing Values =====")
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].mean())
df['Price'] = df['Price'].fillna(df['Price'].mean())
```

```python
print("\n===== Q4: Drop Missing Category/Region =====")
df = df.dropna(subset=['Product Category', 'Region'])
```

```python
print("\n===== Q5: Create Revenue Column =====")
df['Revenue'] = df['Quantity'] * df['Price']
print(df.head())
```

```python
print("\n===== Q6: Total Revenue =====")
total_revenue = np.sum(df['Revenue'])
print("Total Revenue:", total_revenue)
```

```python
print("\n===== Q7: Revenue by Product Category =====")
category_rev = df.groupby('Product Category')['Revenue'].sum()
print(category_rev)
```

```python
print("\n===== Q8: Top & Bottom Categories =====")
top3 = category_rev.sort_values(ascending=False).head(3)
bottom3 = category_rev.sort_values().head(3)
print("Top 3:\n", top3)
print("Bottom 3:\n", bottom3)
```

```python

print("\n===== Q9: Revenue by Region =====")
region_rev = df.groupby('Region')['Revenue'].sum()
print(region_rev)
print("Highest Revenue Region:", region_rev.idxmax())
print("Lowest Revenue Region:", region_rev.idxmin())
```

```python
print("\n===== Q10: Monthly Analysis =====")
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month
```

```python
monthly_rev = df.groupby('Month')['Revenue'].sum()
print("Monthly Revenue:\n", monthly_rev)

print("Mean Revenue:", np.mean(df['Revenue']))
print("Median Revenue:", np.median(df['Revenue']))
print("Standard Deviation:", np.std(df['Revenue']))
```
